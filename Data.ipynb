```python
import requests
import pandas as pd
from pandas import DataFrame
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
url = "http://tour-pedia.org/api/getPlaces?category=accommodation&location=Tuscany"
resp = requests.get(url)
data = resp.json()
df = DataFrame(data)
df = df[(df.numReviews > 0) & (~(df.subCategory.isna())) & (~(df.polarity.isna()))]

#ACCOMMODATION!!!
###MOTEL
b_motel = df[df.subCategory.str.lower().str.contains('motel')]
b_motel['Sub']='Motel'

new = []
for index, row in b_motel.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_motel['vader'] = new

##HOTEL
b_hotel = df[df.subCategory.str.lower().str.contains('hotel')]
b_hotel['Sub']='Hotel'
new = []
for index, row in b_hotel.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_hotel['vader'] = new

####HOSTEL
b_hostel = df[df.subCategory.str.lower().str.contains('hostel')]
b_hostel['Sub']='Hostel'

new = []
for index, row in b_hostel.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_hostel['vader'] = new


###RESORT
b_resort = df[df.subCategory.str.lower().str.contains('resort')]
b_resort['Sub']='Resort'

new = []
for index, row in b_resort.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_resort['vader'] = new

###BED&BREAKFAST
b_bnb = df[df.subCategory.str.lower().str.contains('bed & breakfast')]
b_bnb['Sub']='Bed & Breakfast'

new = []
for index, row in b_bnb.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_bnb['vader'] = new


url = "http://tour-pedia.org/api/getPlaces?category=attraction&location=Tuscany"
resp = requests.get(url)
data = resp.json()

df = DataFrame(data)
df = df[(df.numReviews > 0) & (~(df.subCategory.isna())) & (~(df.polarity.isna()))]


##NIGHTLIFE!!!!

b_bar = df[df.subCategory.str.lower().str.contains('bar')]
b_bar['Sub']='Nightlife'
#BAR
new = []
for index, row in b_bar.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_bar['vader'] = new


###BEER
b_beer = df[df.subCategory.str.lower().str.contains('beer')]
b_beer['Sub']='Nightlife'

new = []
for index, row in b_beer.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_beer['vader'] = new

###CLUB
b_club = df[df.subCategory.str.lower().str.contains('club')]
b_club['Sub']='Nightlife'

new = []
for index, row in b_club.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_club['vader'] = new

#PUB
b_pub = df[df.subCategory.str.lower().str.contains('pub')]
b_pub['Sub']='Nightlife'

new = []
for index, row in b_pub.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_pub['vader'] = new

#NIGHLIFE
b_nightlife = df[df.subCategory.str.lower().str.contains('nightlife')]
b_nightlife['Sub']='Nightlife'

new = []
for index, row in b_nightlife.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_nightlife['vader'] = new

#CASINO
b_casino = df[df.subCategory.str.lower().str.contains('casino')]
b_casino['Sub']='Nightlife'

new = []
for index, row in b_casino.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_casino['vader'] = new


###religious
#church
b_church = df[df.subCategory.str.lower().str.contains('church')]
b_church['Sub']='Religious'

new = []
for index, row in b_church.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_church['vader'] = new

#synagogue
b_synagogue = df[df.subCategory.str.lower().str.contains('synagogue')]
b_synagogue['Sub']='Religious'

new = []
for index, row in b_synagogue.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_synagogue['vader'] = new

#mosque
b_mosque = df[df.subCategory.str.lower().str.contains('mosque')]
b_mosque['Sub']='Religious'

new = []
for index, row in b_mosque.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_mosque['vader'] = new

#spiritual centre
b_spiritual = df[df.subCategory.str.lower().str.contains('spiritual center')]
b_spiritual['Sub']='Religious'

new = []
for index, row in b_spiritual.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_spiritual['vader'] = new


#ARTS
b_art = df[df.subCategory.str.lower().str.contains('art')]
b_art['Sub']='Arts'

new = []
for index, row in b_art.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_art['vader'] = new

##music
b_music = df[df.subCategory.str.lower().str.contains('music')]
b_music['Sub']='Arts'

new = []
for index, row in b_music.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_music['vader'] = new

#theater
b_theater = df[df.subCategory.str.lower().str.contains('theater')]
b_theater['Sub']='Arts'

new = []
for index, row in b_theater.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_theater['vader'] = new

#gallery
b_gallery = df[df.subCategory.str.lower().str.contains('gallery')]
b_gallery['Sub']='Arts'

new = []
for index, row in b_gallery.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_gallery['vader'] = new

#auditorium
b_auditorium = df[df.subCategory.str.lower().str.contains('auditorium')]
b_auditorium['Sub']='Arts'

new = []
for index, row in b_auditorium.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_auditorium['vader'] = new

##SPORTS 
b_athletics = df[df.subCategory.str.lower().str.contains('athletics')]
b_athletics['Sub']='Sports'

new = []
for index, row in b_athletics.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_athletics['vader'] = new

##cricket
b_cricket = df[df.subCategory.str.lower().str.contains('cricket')]
b_cricket['Sub']='Sports'

new = []
for index, row in b_cricket.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_cricket['vader'] = new

#tennis
b_tennis = df[df.subCategory.str.lower().str.contains('tennis')]
b_tennis['Sub']='Sports'

new = []
for index, row in b_tennis.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_tennis['vader'] = new

#soccer
b_soccer = df[df.subCategory.str.lower().str.contains('soccer')]
b_soccer['Sub']='Sports'

new = []
for index, row in b_soccer.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_soccer['vader'] = new

#basketball
b_basketball = df[df.subCategory.str.lower().str.contains('basketball')]
b_basketball['Sub']='Sports'

new = []
for index, row in b_basketball.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_basketball['vader'] = new

#ski
b_ski = df[df.subCategory.str.lower().str.contains('ski')]
b_ski['Sub']='Sports'

new = []
for index, row in b_ski.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_ski['vader'] = new

#skatin
b_skating = df[df.subCategory.str.lower().str.contains('skating')]
b_skating['Sub']='Sports'

new = []
for index, row in b_skating.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_skating['vader'] = new

#hockey
b_hockey = df[df.subCategory.str.lower().str.contains('hockey')]
b_hockey['Sub']='Sports'

new = []
for index, row in b_hockey.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_hockey['vader'] = new

#bowling
b_bowling = df[df.subCategory.str.lower().str.contains('bowling')]
b_bowling['Sub']='Sports'

new = []
for index, row in b_bowling.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_bowling['vader'] = new

#golf
b_golf = df[df.subCategory.str.lower().str.contains('golf')]
b_golf['Sub']='Sports'

new = []
for index, row in b_golf.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_golf['vader'] = new

#football
b_football = df[df.subCategory.str.lower().str.contains('football')]
b_football['Sub']='Sports'

new = []
for index, row in b_football.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_football['vader'] = new

###RECREATIONAL
b_park = df[df.subCategory.str.lower().str.contains('park')]
b_park['Sub']='Recreational'

new = []
for index, row in b_park.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_park['vader'] = new

#zoo
b_zoo = df[df.subCategory.str.lower().str.contains('zoo')]
b_zoo['Sub']='Recreational'

new = []
for index, row in b_zoo.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_zoo['vader'] = new

#lake
b_lake = df[df.subCategory.str.lower().str.contains('lake')]
b_lake['Sub']='Recreational'

new = []
for index, row in b_lake.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_lake['vader'] = new

#garden
b_garden = df[df.subCategory.str.lower().str.contains('garden')]
b_garden['Sub']='Recreational'

new = []
for index, row in b_garden.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_garden['vader'] = new

#plaza
b_plaza = df[df.subCategory.str.lower().str.contains('plaza')]
b_plaza['Sub']='Recreational'

new = []
for index, row in b_plaza.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_plaza['vader'] = new

#mall
b_mall = df[df.subCategory.str.lower().str.contains('mall')]
b_mall['Sub']='Recreational'

new = []
for index, row in b_mall.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_mall['vader'] = new

#market
b_market = df[df.subCategory.str.lower().str.contains('market')]
b_market['Sub']='Recreational'

new = []
for index, row in b_market.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_market['vader'] = new

#aquarium
b_aquarium = df[df.subCategory.str.lower().str.contains('aquarium')]
b_aquarium['Sub']='Recreational'

new = []
for index, row in b_aquarium.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_aquarium['vader'] = new

#arcade
b_arcade = df[df.subCategory.str.lower().str.contains('arcade')]
b_arcade['Sub']='Recreational'

new = []
for index, row in b_arcade.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_arcade['vader'] = new

#beach
b_beach = df[df.subCategory.str.lower().str.contains('beach')]
b_beach['Sub']='Recreational'

new = []
for index, row in b_beach.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_beach['vader'] = new

#castle
b_castle = df[df.subCategory.str.lower().str.contains('castle')]
b_castle['Sub']='Recreational'

new = []
for index, row in b_castle.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_castle['vader'] = new

#city
b_city = df[df.subCategory.str.lower().str.contains('city')]
b_city['Sub']='Recreational'

new = []
for index, row in b_city.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_city['vader'] = new

url = "http://tour-pedia.org/api/getPlaces?category=restaurant&location=Tuscany"
resp = requests.get(url)
data = resp.json()

df = DataFrame(data)
df = df[(df.numReviews > 0) & (~(df.subCategory.isna())) & (~(df.polarity.isna()))]

###FAST FOOD
b_pizza = df[df.subCategory.str.lower().str.contains('pizza')]
b_pizza['Sub']='Fast Food'
new = []
for index, row in b_pizza.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_pizza['vader'] = new

#burger
b_burger = df[df.subCategory.str.lower().str.contains('burger')]
b_burger['Sub']='Fast Food'
new = []
for index, row in b_burger.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_burger['vader'] = new

#diner
b_diner = df[df.subCategory.str.lower().str.contains('diner')]
b_diner['Sub']='Fast Food'
new = []
for index, row in b_diner.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_diner['vader'] = new

#snack
b_snack = df[df.subCategory.str.lower().str.contains('snack')]
b_snack['Sub']='Fast Food'
new = []
for index, row in b_snack.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_snack['vader'] = new

#foodcourt
b_foodcourt = df[df.subCategory.str.lower().str.contains('food court')]
b_foodcourt['Sub']='Fast Food'
new = []
for index, row in b_foodcourt.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_foodcourt['vader'] = new

#sandwich
b_sandwich = df[df.subCategory.str.lower().str.contains('sandwich')]
b_sandwich['Sub']='Fast Food'
new = []
for index, row in b_sandwich.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_sandwich['vader'] = new

#bbq
b_bbq = df[df.subCategory.str.lower().str.contains('bbq')]
b_bbq['Sub']='Fast Food'

new = []
for index, row in b_bbq.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_bbq['vader'] = new

#foodtruck
b_foodtruck = df[df.subCategory.str.lower().str.contains('food truck')]
b_foodtruck['Sub']='Fast Food'
new = []
for index, row in b_foodtruck.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_foodtruck['vader'] = new

#cafeteria
b_cafeteria = df[df.subCategory.str.lower().str.contains('cafeteria')]
b_cafeteria['Sub']='Fast Food'
new = []
for index, row in b_cafeteria.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_cafeteria['vader'] = new

#deli
b_deli = df[df.subCategory.str.lower().str.contains('deli')]
b_deli['Sub']='Fast Food'
new = []
for index, row in b_deli.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_deli['vader'] = new


#chicken
b_chicken = df[df.subCategory.str.lower().str.contains('chicken')]
b_chicken['Sub']='Fast Food'
new = []
for index, row in b_chicken.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_chicken['vader'] = new


#fishchips
b_fishchips = df[df.subCategory.str.lower().str.contains('fish & chips')]
b_fishchips['Sub']='Fast Food'
new = []
for index, row in b_fishchips.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_fishchips['vader'] = new



###RESTAURANT
b_restaurant = df[df.subCategory.str.lower().str.contains('restaurant')]
b_restaurant['Sub']='Restaurant'
new = []
for index, row in b_restaurant.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_restaurant['vader'] = new

#steakhouse
b_steakhouse = df[df.subCategory.str.lower().str.contains('steakhouse')]
b_steakhouse['Sub']='Restaurant'
new = []
for index, row in b_steakhouse.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_steakhouse['vader'] = new


#ramen
b_ramen = df[df.subCategory.str.lower().str.contains('ramen')]
b_ramen['Sub']='Restaurant'
new = []
for index, row in b_ramen.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_ramen['vader'] = new

#salad
b_salad = df[df.subCategory.str.lower().str.contains('salad')]
b_salad['Sub']='Restaurant'
new = []
for index, row in b_salad.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_salad['vader'] = new

#gourmet
b_gourmet = df[df.subCategory.str.lower().str.contains('gourmet')]
b_gourmet['Sub']='Restaurant'
new = []
for index, row in b_gourmet.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_gourmet['vader'] = new


###BAKERY
b_bakery = df[df.subCategory.str.lower().str.contains('bakery')]
b_bakery['Sub']='Bakery'
new = []
for index, row in b_bakery.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_bakery['vader'] = new

#breakfast
b_breakfast = df[df.subCategory.str.lower().str.contains('breakfast')]
b_breakfast['Sub']='Bakery'

new = []
for index, row in b_breakfast.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_breakfast['vader'] = new

#bagel
b_bagel = df[df.subCategory.str.lower().str.contains('bagel')]
b_bagel['Sub']='Bakery'
new = []
for index, row in b_bagel.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_bagel['vader'] = new



###DESSERT
b_icecream = df[df.subCategory.str.lower().str.contains('ice cream')]
b_icecream['Sub']='Dessert'
new = []
for index, row in b_icecream.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_icecream['vader'] = new

#dessert
b_dessert = df[df.subCategory.str.lower().str.contains('dessert')]
b_dessert['Sub']='Dessert'

new = []
for index, row in b_dessert.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_dessert['vader'] = new



##CAFE
b_coffee = df[df.subCategory.str.lower().str.contains('coffee')]
b_coffee['Sub']='Cafe and Drinks'
new = []
for index, row in b_coffee.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_coffee['vader'] = new


#cafe
b_caf = df[df.subCategory.str.lower().str.contains('caf')]
b_caf['Sub']='Cafe and Drinks'
new = []
for index, row in b_caf.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_caf['vader'] = new

#tea
b_tea = df[df.subCategory.str.lower().str.contains('tea')]
b_tea['Sub']='Cafe and Drinks'
new = []
for index, row in b_tea.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_tea['vader'] = new

###POI
url = "http://tour-pedia.org/api/getPlaces?category=poi&location=Tuscany"
resp = requests.get(url)
data = resp.json()

df = DataFrame(data)
df = df[(df.numReviews > 0) & (~(df.subCategory.isna())) & (~(df.polarity.isna()))]

#uni
b_uni = df[df.subCategory.str.lower().str.contains('university')]
b_uni['Sub']='University'
new = []
for index, row in b_uni.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_uni['vader'] = new

#library
b_lib = df[df.subCategory.str.lower().str.contains('library')]
b_lib['Sub']='Library'
new = []
for index, row in b_lib.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_lib['vader'] = new


#museum
b_museum = df[df.subCategory.str.lower().str.contains('museum')]
b_museum['Sub']='Museum'
new = []
for index, row in b_museum.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_museum['vader'] = new

#monument
b_monument = df[df.subCategory.str.lower().str.contains('monument')]
b_monument['Sub']='Monument'
new = []
for index, row in b_monument.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_monument['vader'] = new

#conv
b_conv = df[df.subCategory.str.lower().str.contains('convention center')]
b_conv['Sub']='Convention Center'
new = []
for index, row in b_conv.iterrows():
    url = requests.get('http://tour-pedia.org/api/getReviewsByPlaceId?placeId={}'.format(row['id']))
    data = url.json()
    text = [item for item in data if item.get('language')=='en']
    text = [li['text'] for li in text]
    
    sent = 0.0
    count = 0
    for sentence in text:
        count = len(text)
        vs = analyzer.polarity_scores(sentence)
        sent += vs['compound']
        average = round((sent/count), 2)
        t = average
    new.append(t)
b_conv['vader'] = new

frame=[b_bar,b_beer,b_club,b_pub,b_nightlife,b_casino,b_hotel,b_resort, b_bnb, b_hostel, b_motel, b_church,b_synagogue,b_mosque,b_spiritual,
         b_theater,b_music,b_art,b_gallery,b_auditorium,
         b_athletics,b_cricket,b_tennis,b_soccer,b_basketball,b_ski,b_skating,b_hockey,b_bowling,b_golf,b_football,
         b_park,b_zoo,b_lake,b_garden,b_plaza,b_mall,b_market,b_aquarium,b_arcade,b_beach,b_castle,b_city,
         b_uni,b_lib,b_museum,b_monument,b_conv,
         b_pizza,b_burger,b_diner,b_snack,b_foodcourt,b_sandwich,b_bbq,b_foodtruck,b_cafeteria,b_deli,b_chicken,b_fishchips,
         b_restaurant,b_steakhouse,b_ramen,b_salad,b_gourmet,
         b_coffee,b_caf,b_tea,
         b_bakery,b_breakfast,b_bagel,
         b_icecream,b_dessert]

result = pd.concat(frame).drop_duplicates().reset_index(drop=True)
result.to_csv('tuscany.csv', index = False) 
sort=True
```
